```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create data frame from H-alpha csv

#csv file
h_alpha_file = "./HI to H-alpha Classification - Sheet1.csv"

#initialize dataframe with csv
h_alpha_df = pd.read_csv(h_alpha_file)

print(h_alpha_df.columns)
```

```python
#create histogram

#get unique galaxies from data frame
galaxies = pd.unique(h_alpha_df["Galaxy"])

#loop through galaxies and determine detection
for galaxy in galaxies:

    #initilize detection count
    detection = 0
    no_detection = 0

    #iterate over rows in csv
    for index, row in enumerate(h_alpha_df["Galaxy"]):

        #check if galaxy is unique
        if galaxy == row:

            #set detection variable
            detected = h_alpha_df["Classification"][index]
            #print(f"{galaxy}", detected)

            #check if detected
            if detected != "No Coverage" and detected != "No Emission":

                #if detected increase detection
                detection += 1

            else:
                #if not detected increase no_detection
                no_detection += 1

    #plot histogram for unique galaxy
    plt.figure(figsize=(10,10))
    plt.bar("Detections", detection, color='g', width=1)
    plt.bar("Non-Detections", no_detection, color='r', width=1)
    plt.title(f"Detections vs. Non-Detections in Galaxy: {galaxy}")
    plt.ylabel("Count")
    plt.show()
                
```
