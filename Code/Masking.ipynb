```python
#import libraries
import numpy as np
import astropy.units as u
from astropy.coordinates import SkyCoord
import pyregion
from astroquery.gaia import Gaia  
import sep
```

```python
#query structure
cone_query_base = '''
SELECT
COUNT(source_id)
FROM gaiadr2.gaia_source
WHERE 1=CONTAINS(
  POINT(ra, dec),
  CIRCLE({ra}, {dec}, {deg}))
'''
```

```python
#file info
region_file = "..//Data/STARTAstro_dwarf_shells/regions/ddo50.reg"
```

```python
#load regions from region file
regions = pyregion.open(region_file)

#iterate through regions
for region in regions:

    #check if applicable region
    if region.name != "ellipse":
        continue
        
    #get coords for region
    ra, dec, a_arcsec, b_arcsec, pa_deg = region.coord_list

    #convert arcsecs to degrees
    deg = a_arcsec / 3600

    #format cone query
    cone_query = cone_query_base.format(ra=ra, dec=dec, deg=1)

    #print(cone_query)

    #query gaia
    cone_job = Gaia.launch_job(cone_query)

    #get query results
    cone_results = cone_job.get_results()

    print(cone_results)
```
