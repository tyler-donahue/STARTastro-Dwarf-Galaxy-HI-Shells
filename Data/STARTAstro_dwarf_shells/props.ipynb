```python
import pandas as pd
import os
```

```python
# --- Input files ---
holes = "shells_LT_pokhrel.txt"
sample = "sample.txt"
```

```python
# --- Read data ---
holes_df = pd.read_csv(holes, sep=r"\t+", engine="python", comment="#")
dist_df = pd.read_csv(sample, sep="\t", comment="#")
print(holes_df.columns)
merged = holes_df.merge(dist_df[["Galaxy", "Dist"]], on="Galaxy", how="left")
```

```python
# Filter rows with Type 3 holes
type3_holes = merged[merged["Type"] == 3]

# Count Type 3 holes per galaxy
type3_counts = type3_holes.groupby("Galaxy").size().reset_index(name="Type3_count")

# Keep only galaxies with at least one Type 3 hole
type3_counts = type3_counts[type3_counts["Type3_count"] > 0]

# Filter rows with Type 3 holes
type2_holes = merged[merged["Type"] == 2]

# Count Type 3 holes per galaxy
type2_counts = type2_holes.groupby("Galaxy").size().reset_index(name="Type2_count")

type2_counts = type2_counts[type2_counts["Type2_count"] > 0]

print(type3_counts,  type2_counts)
```
