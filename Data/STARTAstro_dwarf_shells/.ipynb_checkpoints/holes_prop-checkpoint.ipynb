```python
from astropy.io import fits
from astropy.wcs import WCS
from photutils.aperture import EllipticalAperture
from photutils.aperture import EllipticalAnnulus
from photutils.aperture import aperture_photometry
from astropy.coordinates import SkyCoord
import astropy.units as u
import pyregion
import numpy as np
from photutils.centroids import centroid_com
from skimage.measure import regionprops, label
from astropy.nddata import Cutout2D
import matplotlib.pyplot as plt
import statmorph
from photutils.segmentation import detect_sources

from matplotlib.patches import Ellipse
from astropy.visualization import simple_norm

import warnings
warnings.simplefilter('ignore')
```

```python
# Load image
hdu = fits.open('/Users/mansi/Research/dwarf_shells/data/wlm_ha.fits')
header = hdu[0].header
wcs = WCS(header)
arcsec_per_pixel = np.abs(hdu[0].header["CD2_2"]) * 3600
unit = 10**-11.6749 / 1539

regions = pyregion.open('WLM.reg')

apertures = []

for reg in regions:
    if reg.name != "ellipse":
        continue

    # coord_list: [RA, Dec, a_arcsec, b_arcsec, PA]
    ra, dec, a_arcsec, b_arcsec, pa_deg = reg.coord_list

    # Convert center to pixel
    skycoord = SkyCoord(ra=ra*u.deg, dec=dec*u.deg)
    x, y = skycoord.to_pixel(wcs)

    # Convert arcsec → pixels
    a_pix = a_arcsec* 3600 / arcsec_per_pixel
    b_pix = b_arcsec* 3600 / arcsec_per_pixel
    theta = np.deg2rad(pa_deg)

    if(x>0 and y>0):

        aperture = EllipticalAperture((x, y), a_pix, b_pix, theta)
        phot_table = aperture_photometry(hdu[0].data*unit, aperture)
        print(phot_table['aperture_sum'])

        core_ap = EllipticalAperture((x, y), 0.5*a_pix, 0.5*b_pix, theta)
        rim_ap = EllipticalAnnulus((x, y), 0.8*a_pix, 2*a_pix, 2*b_pix, 0.8*b_pix, theta)
        core_flux = aperture_photometry(hdu[0].data*unit, core_ap)['aperture_sum'][0]
        rim_flux = aperture_photometry(hdu[0].data*unit, rim_ap)['aperture_sum'][0]
        rim_to_core = rim_flux / core_flux
        
        if(rim_to_core>10):

            cutout_obj = Cutout2D(data=hdu[0].data*unit, position=(x,y), size=5*a_pix)
            cutout = cutout_obj.data 
            
            # Shift aperture center to cutout frame
            x_c, y_c = cutout_obj.input_position_cutout  # position of (x, y) in cutout pixel coordinates
            
            # Plot
            plt.figure(figsize=(5, 5))
            plt.imshow(np.log10(cutout), origin='lower', cmap='magma')
            plt.title(f'Rim/Core = {rim_to_core:.2f} at x={x:.1f}, y={y:.1f}')
            
            # Plot core ellipse
            core_ellipse = Ellipse((x_c, y_c), width=a_pix, height=b_pix,
                                   angle=np.degrees(theta), edgecolor='lime', facecolor='none', lw=2)
            plt.gca().add_patch(core_ellipse)
            
            # Plot rim ellipses
            rim_outer = Ellipse((x_c, y_c), width=2*a_pix, height=2*b_pix,
                                angle=np.degrees(theta), edgecolor='cyan', facecolor='none', linestyle='--', lw=2)
            rim_inner = Ellipse((x_c, y_c), width=0.8*a_pix*2, height=0.8*b_pix*2,
                                angle=np.degrees(theta), edgecolor='cyan', facecolor='none', linestyle='--', lw=2)
            
            plt.gca().add_patch(rim_outer)
            plt.gca().add_patch(rim_inner)
            
            plt.axis('off')
            plt.tight_layout()
            plt.show()
            
            # # Create segmentation map
            # threshold = np.percentile(cutout, 90)
            # segm = detect_sources(cutout, threshold, npixels=8)
            # segmap = (segm.data == 1).astype(int)  # assumes 1 object
            
            # # Run statmorph
            # morph_list = statmorph.source_morphology(cutout, segmap, gain=1.0)
            # print(morph_list)
            # morph = morph_list[0]
            # # Step 4: Access parameters
            # print("Elongation:", morph.elongation_centroid)
            # print("Asymmetry:", morph.asymmetry)
            # print("Gini:", morph.gini)
            # print("Centroid (x, y):", morph.xc_centroid, morph.yc_centroid)
            # print("Sérsic n:", morph.sersic_n)
            # from statmorph.utils.image_diagnostics import make_figure
            # fig = make_figure(morph)

```
